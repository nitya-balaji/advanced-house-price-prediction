```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

#load the datasets you saved
train=pd.read_csv("X_train.csv")
test=pd.read_csv("X_test.csv")

#define the features we selected
#this makes sure we use the same 20 columns for training and testing
features=[col for col in train.columns if col not in ["Id","SalePrice"]]
X_train=train[features]
y_train=train["SalePrice"]
X_test=test[features]

#initialize the model
model=RandomForestRegressor(n_estimators=100,random_state=42)

#fit the model
model.fit(X_train,y_train)

#make predictions
#important: we use exp because saleprice was log transformed earlier
log_predictions=model.predict(X_test)
final_predictions=np.exp(log_predictions)

#create the submission dataframe
submission=pd.DataFrame({
    "Id":test["Id"],
    "SalePrice":final_predictions
})

#save to csv for kaggle
submission.to_csv("submission.csv",index=False)
print("submission file is ready!")
print(submission.head()) #look at first few rows of final submission file
```
