```python
#Data Analysis Phase 
#Main aim: To understand more about the data we are working with
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#display all columns of the dataframe 
pd.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv("train.csv")

#print shape of the dataset with rows and columns
print(dataset.shape)
```

```python
#print the top 5 records (all columns will be shown + first 5 rows due to set_option function)
dataset.head() 
```

```python
#here we will check the percentage of non values/missing values are present in each feature/column
#step 1: make the list of features which has missing values
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

#step 2: print the feature name and percentage of missing values
for feature in features_with_na: #list from previous step
    print(feature, np.round(dataset[feature].isnull().mean()*100, 4), "% missing values") # .mean() = number of missing values in column/total items in the column
```

```python
#finding the relationship btwn missing values and sales price through visualization
for feature in features_with_na:
    data = dataset.copy()

    #making a variable that indicates 1 if the observation was missing or 0 otherwise 
    data[feature] = np.where(data[feature].isnull(), 1, 0)

    #calculate the mean SalePrice where the information is missing or present (doing this to look at how the number of missing values will affect the sales price)
    data.groupby(feature)["SalePrice"].median().plot.bar(color=["navy", "orangered"] ) #grouping features with missing info and features with present info separately, finding the median of their house prices, and plotting them in a bar graph
    plt.title(feature)
    plt.show()

```

Based on the above visualizations, it is clear that there is a relation between the missing values and the dependent variable. Due to this, the NaN values (also known as the missing values) need to be replaced with something meaningful, which will be explored in the Feature Engineering portion of this project. 

```python
#from our dataset, some features (e.g. Id) is not required (meaning that Id number has no relationship with the sale price)
print("Id of Houses {}".format(len(dataset.Id))) #before cleaning the dataset, we need to keep track of how many houses are in the dataset
```

```python
#finding out how many features are numerical variables
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != "O"] #O = object (aka string field values)
print("Number of numerical variables: ", len(numerical_features))

#visualize the numerical variables
dataset[numerical_features].head() #display all numerical features 
```

Analysis of Temporal Variables (Year/Datetime Variables): Our dataset contains 4 features related to years (e.g., Year Built, Year Sold). However, the raw year (like 1990) isn't as useful to a model as Age. In the Feature Engineering phase, we will transform these dates into "Time Elapsed" by calculating the difference between the Year Sold and the other year variables. This tells us exactly how old the house or its features were at the time of sale, which has a much stronger correlation with the price.

```python
#finding the list of variables that contain year information
year_feature = [feature for feature in numerical_features if "Yr" in feature or "Year" in feature]
print(year_feature)
```

```python
#looking at the content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique()) #.unique() allows us to view all unique values (no values will be repeated)
```

```python
#analyzing the temporal datetime variables 
#we will check whether there is a relation btwn year the house is sold and the sale price
dataset.groupby("YrSold")["SalePrice"].median().plot() #taking the median sale price for each unique group/YrSold value and plotting that
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price vs. Year Sold")
```

We can see from the above visualization that the house price decrease as the year sold increases. Since this isn't a clear conclusion that can be drawn just yet, we can transforming raw Year dates into Age by subtracting them from the YrSold. This allows us to see the direct relationship between the age of a feature (like the garage or the house itself) and the final SalePrice. 

```python
#comparing the difference between ALL years feature with YrSold 
for feature in year_feature:
    if feature != "YrSold": #can't subtract year sold from itself (won't provide useful info)
        data = dataset.copy() #copying the original dataset to be able to look at all column/feature values along with sale price
        #we are capturing the difference btwn year variable and the year the house was sold for - this let's us get the "age" difference and make better observations with how sale prices are affected
        data[feature] = data["YrSold"] - data[feature] #represents the age difference
    
        #plot each age difference along with sale price
        plt.scatter(x=data[feature], y=data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.show()
        
```

```python
#looking at how many discrete variables we have 
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_feature+["Id"]] #setting a length of 25 to be our threshold to identify discrete feature
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
#looking at the specific discrete features
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
#finding relationship btwn discrete features and sales price
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)["SalePrice"].median().plot.bar() #grouping all the different values for each discrete feature, finding the median of each interval's saleprice, and then plotting these values in a bar plot
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

```python
#looking at how many continuous features we have 
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+["Id"]]
print("Continuous feature Count {}".format(len(continuous_feature)))
continuous_feature
```

```python
#analyzing the continuous values by creating histrograms to understand the distribution of each feature (can't just plot scatterplots comparing sales price for this reason)
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count") #aka frequency 
    plt.title(feature)
    plt.show()
```

Analysis of Continuous Feature Distributions:
Looking at the histograms above, it is clear that many features do not follow a Gaussian (Normal) Distribution. Instead, they are skewed (they have long tails). In the Feature Engineering phase of this project, we will apply logarithmic transformations to handle this skewness and bring the data closer to a Gaussian distribution. This is essential because many machine learning algorithms perform better when the numerical input features are normally distributed.

```python
#using logarithmic transformation 
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique(): #checking if any unique value within each continuous feature is 0 - if so we need to pass (bc log function can't take in an input of 0)
        pass
    else:
        data[feature]=np.log(data[feature]) #log transformation applied
        data["SalePrice"]=np.log(data["SalePrice"]) #need to apply log transformation to sale price because sale price itself is highly skewed 
        plt.scatter(x=data[feature],y=data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(feature)
        plt.show()
```

```python
#looking at outliers
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature) #finding outliers using box plot 
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
